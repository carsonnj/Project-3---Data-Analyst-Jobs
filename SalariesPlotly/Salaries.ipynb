```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pathlib import Path
import plotly.graph_objects as go
import json
import matplotlib.ticker as mticker
```

```python
salary_df= pd.read_csv("../Resources/Salary_Data_Cleaned.csv")
salary_df
```

```python
years_of_exp = salary_df.groupby('Years of Experience')['Salary'].mean().reset_index()
years_of_exp['Salary'] = years_of_exp['Salary'].round(0).astype(int)

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(years_of_exp['Years of Experience'], years_of_exp['Salary'], marker='o', color='blue', linestyle='-')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.title('Average Salary per Years of Experience (Line Chart)')
plt.grid(True)
plt.show()
```

```python
json_data = years_of_exp.to_json(orient='records')

# Save the JSON data to a file
with open('years_exp.json', 'w') as json_file:
    json_file.write(json_data)
years_of_exp
```

```python
education_level = salary_df.groupby('Education Level')['Salary'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(education_level['Education Level'], education_level['Salary'], color='blue')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Education Level')
plt.show()
```

```python
json_data = education_level.to_json(orient='records')

# Save the JSON data to a file
with open('education_level.json', 'w') as json_file:
    json_file.write(json_data)
    
education_level
```
